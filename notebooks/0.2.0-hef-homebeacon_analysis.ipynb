```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
b7 = pd.DataFrame()
for file in os.listdir('/Users/hagenfritz/Projects/utx000/data/raw/bpeace2/beacon/B07/sensirion/'):
    temp = pd.read_csv(f'/Users/hagenfritz/Projects/utx000/data/raw/bpeace2/beacon/B07/sensirion/{file}',
                      index_col=0,parse_dates=True,infer_datetime_format=True)
    b7 = pd.concat([b7,temp])
    
b7 = b7.resample('5T').mean()
```
