# Individual-Level Inspection
We want to look at a few individuals to get a sense of the data and what gaps are present. The entire study period is also quite so we can restrict it to just a month.

````python
gps['Beiwe'].unique()
````

Choosing randomly from the list above (first, last, and somewhere in the middle), we can inspect the GPS data.

````python
for coord in ['Lat','Long']:
    print(f'Looking at {coord}:')
    for pt in ['2xtqkfz1','xxvnhauv','itmylz3g']:
        fig, ax = plt.subplots(figsize=(16,4))
        # gps data per individual
        gps_pt = gps[gps['Beiwe'] == pt]
        # restricting again for the date
        #gps_pt = gps_pt[datetime(2020,6,7):datetime(2020,6,9)]
        # info per individual
        info_pt = info[info['Beiwe'] == pt]

        ax.scatter(gps_pt.index,gps_pt[coord], color='black', s=5)
        ax.axhline(info_pt[coord].values[0],color='red',linestyle='dashed')

        plt.show()
        plt.close()
````
